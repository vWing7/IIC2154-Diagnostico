```python
import pandas as pd
import re
```

## Cleaning

```python
twitter_df = pd.read_json("farmers-protest-tweets-2021-03-5.json", lines=True)
```

```python
twitter_df
```

```python
user_series = twitter_df["user"].apply(lambda x: x["username"])
```

```python
twitter_df["username"] = user_series
```

## Functions

```python
def most_retweeted(dataframe):
  top_10 = dataframe.nlargest(10, "retweetCount")
  return top_10
```

```python
def user_most_tweets(dataframe):
  grouped_df = dataframe.groupby(["username"]).count()
  top_10 = grouped_df.nlargest(10, "content")
  top_10.rename(columns = {"url": "tweetCount"}, inplace=True)
  return top_10["tweetCount"]
```

```python
def days_most_tweets(dataframe):
  grouped_df = dataframe.groupby([dataframe["date"].dt.date]).count()
  top_10 = grouped_df.nlargest(10, "content")
  top_10.rename(columns = {"url": "tweetCount"}, inplace=True)
  return top_10["tweetCount"]
```

```python
#https://stackoverflow.com/questions/2527892/parsing-a-tweet-to-extract-hashtags-into-an-array
#https://stackoverflow.com/questions/55672912/faster-way-to-count-total-occurrences-of-values-in-a-column-of-lists-in-pandas

def most_used_hashtag(dataframe):
  comment_series = dataframe["content"]
  all_hashtags = comment_series.str.findall("#(\w+)")
  hashtag_count = pd.DataFrame(all_hashtags.values.tolist()).stack().value_counts()
  return hashtag_count.head(n=10)
```

```python
def main(df, info_required):
  if info_required == 0:
    return most_retweeted(df)
  elif info_required == 1:
    return user_most_tweets(df)
  elif info_required == 2:
    return days_most_tweets(df)
  elif info_required == 3:
    return most_used_hashtag(df)  
```

## Result

```python
top10_retweets = main(twitter_df, 0)
top10_retweets
```

```python
top10_users = main(twitter_df, 1)
top10_users
```

```python
top10_days = main(twitter_df, 2)
top10_days
```
